```python
from scipy.stats import binom
import scipy as sp
```

```python
for trials in range (0,10000):
    for i in range(0,trials):
        if 1 - binom.cdf(i-1, trials, 0.5) < 0.05:
            power = 1 - binom.cdf(i-1, trials, 0.52)
            print(i , trials, power)
            if power > 0.8 :
                print (power, i , trials)
                break
            else:
                break
```

```python
def prod_two_num(a,b):
    c = a * b
    print(c)

def div_two_num(a,b):
    c = a / b
    print(c)

def prod_div(a,b):
    prod_two_num(a,b)
    div_two_num(a,b)
```

```python
prod_div(6,3)
```

```python
# Function definition is here
def listappend1( mylist ):
    #This changes a passed list into this function
    mylist.append([1,2,3,4])
    print("Values inside the function: ", mylist)
    return

# Now you can call changeme function
mylist = [10,20,30];
listappend1( mylist );
print ("Values outside the function: ", mylist)
```
