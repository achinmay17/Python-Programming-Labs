```python
from scipy.stats import binom
import scipy as sp
```

```python
for trials in range(0,10000):
    for i in range(0,trials):
        if 1 - binom.cdf(i-1,trials,0.5) < 0.05:
            power = 1 - binom.cdf(i-1,trials,0.52)
            print(i,trials,power)
            if power > 0.8:
                print (power,i,trials)
                break
            else:
                break
    
```
